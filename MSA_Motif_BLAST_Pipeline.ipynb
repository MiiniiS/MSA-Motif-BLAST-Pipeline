Performing Multiple Sequence Alignment (MSA)

```python
from Bio. Align import MultipleSeqAlignment
from Bio import AlignIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord

Seq=[SeqRecord(Seq("ATCGGGGTACCCCCCTTTAAA"), id="Seq1"),
     SeqRecord(Seq("ATTTTCCCCCGGGGGGAAATA"), id="Seq1"),
     SeqRecord(Seq("TTTCCGGGAAATTGGCCCCAT"), id ="Seq2")]
Alignment= MultipleSeqAlignment(Seq)
print(Alignment)
```

Searching for Motifs

```python
import re
from Bio.Seq import Seq
Seq=Seq("ATTCTGCGGTTAAGGAAA")
motif="CTG"
matches=[match.start() for match in re.finditer(motif,str(Seq))]
print("Motif Matches:", ", ".join(map(str, matches)))
```

Calculating Molecular Weight

```python
from Bio.SeqUtils import molecular_weight
from Bio.Seq import Seq
Seq=Seq("ATTCTGCGGTTAAGGAAA")
weight=molecular_weight(Seq)
print("Molecular Weight:", weight)
```
